```python
import numpy as np
import matplotlib.pyplot as plt
import pltedit as plte

fig = plt.figure()
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plte.save(fig, "figure.plt")

plt.style.use("ggplot")
fig.show()
```

```python
fig_loaded = plte.load("figure.plt")
fig_styled = plte.set_style(fig_loaded, "seaborn-v0_8")
ax = fig_styled.get_axes()[0]
ax.set_title("Loaded & Styled Figure")
ax.set_xlim(0, 5)
fig_styled
```
